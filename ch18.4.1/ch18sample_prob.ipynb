```python
# We have to create a "symbol" called x
c_list = []

x = Symbol('x')
f = 5*x**2 + 3*x + 5
f_prime = f.diff(x)
f_prime2 = f_prime.diff(x)
f = lambdify(x, f)
f_prime = lambdify(x, f_prime)
f_prime2 = lambdify(x, f_prime2)

print('At a = 0')
x = 1.5
a = 0
tot_list = []
tot = 0
c_list.append(f(a))
c_list.append(f_prime(a))
c_list.append(f_prime2(a))

for i in range(len(c_list)):
    y = c_list[i]*(x - a)**(i)/np.math.factorial(i)
    tot_list.append(y)
    tot += y 
print(tot_list)
print(tot)
print(f(1.5))   
    
```

```python
x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order', 'Ninth Order']

plt.figure(figsize=(10, 8))
for n, label in zip(range(5), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label=label)

plt.plot(x, np.sin(x), 'k', alpha=0.3, label='Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper center',  bbox_to_anchor=(0.25, 0.95), shadow=True)
plt.show()
```

```python
x = Symbol('x')
f = sym.exp(x)
fe = lambdify(x, f)
ter = 15
exp = 0
x = 2
rows = []
for i in range(ter):
    exp = exp + \
       ((x**i)/np.math.factorial(i))
    rows.append([i, exp, np.abs(exp - fe(x))])
pprint(f'Using {i}-terms, {exp}, error of {np.abs(exp - fe(x))}', use_unicode=True)
df_terms = pd.DataFrame(rows, columns=["terms(n)", "f(x)^(n)", "Error"])
print(df_terms)
pprint(f' {f} is:   {(fe(x))}; for x = {x},', use_unicode=True)
```

```python
var( 'x' )
f = sym.exp(x)
a = 0
x0 = 2
n = 9
values = 20
# Create a list of values to evaluate for c in for statement
# to obtain the max value for function (f) evaulated a n+1 degree
cs = [Min(x0,a) + np.abs(x0-a)*i/values for i in range(values+1)]

#The absolute value of f evaulated at n+1 degree for x |f^(n+1)(x)|:
fn = abs(diff(f, x, n+1))
pprint(f'The abs value of the function (f) evaluated at n+1 or {n+1} terms for {fn}.', use_unicode=True)

# Find the maximum M per cs values:
M = Max(*[fn.subs(x,c) for c in cs])
pprint(f'Maximum value for {fn} is {M}.', use_unicode=True)

# Compute the error bound:
En = (M * np.abs(x0-a)**(n+1) / np.math.factorial(n+1))
pprint(f'The Taylor Remainder Estimation at 0  z  {x0}, {En}', use_unicode=True)
```
